```python
# Evaluar sistema con acciones aleatorias
acc=[]
X_ = []
Y_ = []
reward_points = 0#init_reward_points()

for episode in range(10):
    env.reset()
    #print(f'Episodio: {episode}')
    for step in range(10):
        env.render()
        env.step(env.action_space.sample())   # toma una accion aleatoria
        #time.sleep(5)
        reward = 0   #env.compute_reward(env.compute_distance(env.init_reward_points()))
        acc += [reward]
        X_.append(episode)
        Y_.append(np.array(acc).mean())
       
        print('Reecompensa Actual: {0:.3f}'.format(reward))
        #draw_rl_debug(env.step(env.action_space.sample()), reward_points)
    
plt.plot(X_, Y_)
plt.title('Recompensa medida')
plt.xlabel('Numero Episodios')
plt.ylabel('Recompensa')
plt.grid()
plt.show()
env.close()
```
