```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import numpy as np
```

```python
data = pd.read_excel('C:/users/USUARIO/EnvCGRT/Trajectories.xlsx')
```

```python
vx1 = data['x1'].values
vy1 = data['y1'].values
vx2 = data['x2'].values
vy2 = data['y2'].values
vx3 = data['x3'].values
vy3 = data['y3'].values
vx4 = data['x4'].values
vy4 = data['y4'].values
vx5 = data['x5'].values
vy5 = data['y5'].values
```

```python
# Calcular la media de los vectores vx y vy
vx6 = np.mean([vx1, vx2, vx3, vx4, vx5], axis=0)
vy6 = np.mean([vy1, vy2, vy3, vy4, vy5], axis=0)
# Aplicar suavizado con un filtro gaussiano
sigma = 0.85  # Ajusta este valor para ms o menos suavizado
smoothed_vy6 = gaussian_filter1d(vy6, sigma=sigma)

curvas = [(vx1, vy1, 'Inference 1','steelblue', 1.5), (vx2, vy2, 'Inference 2','darkorange', 1.5), 
          (vx3, vy3,'Inference 3','green', 1.5), (vx4, vy4, 'Inference 4','red', 1.5), 
          (vx5, vy5, 'Inference 5','darkviolet', 1.5), (vx6, smoothed_vy6, 'Trajectory reference','black', 2)]

fig, ax = plt.subplots(figsize=(8, 3), dpi=1000)

for vx, vy, label, color, linewidth in curvas:
    ax.plot(vx, vy, label=label, color=color, linewidth=linewidth)

ax.set_title('Trajectory tracking of DRL model',fontsize=8)
ax.set_xlabel('X axis (m)',fontsize=7)
ax.set_ylabel('Y axis (m)',fontsize=7)
ax.tick_params(axis='both', which='major', labelsize=7)
ax.fill_between(x=vx6, y1=smoothed_vy6, y2=0.025, where=smoothed_vy6 > 0.025, interpolate=True, facecolor='green', alpha=.0)
ax.grid(color='grey', ls = '-.', lw = 0.25)
ax.legend(fontsize=6)

x1, x2, y1, y2 = 0.57, 0.59, 0.05, 0.06  # subregion of the original image
axin = ax.inset_axes(
    [0.15, 0.1, 0.25, 0.35],
    xlim=(x1, x2), ylim=(y1, y2))
axin.tick_params(axis='both', which='major', labelsize=5)
axin.grid(color='grey', ls = '-.', lw = 0.2)

for vx, vy, label, color, linewidth in curvas:
    axin.plot(vx, vy, label=label, color=color, linewidth=linewidth)
    
ax.indicate_inset_zoom(axin, edgecolor="black")

plt.ylim(-0.05, 0.08)

plt.show()
```

```python
# Inicializar listas para almacenar las estadsticas
media_x_list = []
media_y_list = []
std_x_list = []
std_y_list = []
correlation_list = []

# Funcin para calcular la correlacin entre dos vectores
def calculate_correlation(vx, vy):
    return np.corrcoef(vx, vy)[0, 1]

# Calcular las estadsticas para cada trayectoria con respecto a la trayectoria promedio suavizada
for vx, vy in zip([vx1, vx2, vx3, vx4, vx5], [vy1, vy2, vy3, vy4, vy5]):
    # Media de la diferencia en X y Y
    media_x = np.mean(vx6 - vx)
    media_y = np.mean(smoothed_vy6 - vy)
    
    # Desviacin estndar de la diferencia en X y Y
    std_x = np.std(vx6 - vx)
    std_y = np.std(smoothed_vy6 - vy)
    
    # Correlacin entre la trayectoria original y la trayectoria promedio suavizada
    correlation = calculate_correlation(vx6 - vx, smoothed_vy6 - vy)
    
    # Almacenar las estadsticas
    media_x_list.append(media_x)
    media_y_list.append(media_y)
    std_x_list.append(std_x)
    std_y_list.append(std_y)
    correlation_list.append(correlation)

# Mostrar los resultados
for i in range(5):
    print(f"Trayectoria {i+1}:")
    print(f"  Media eje X: {media_x_list[i]:.4f}")
    print(f"  Media eje Y: {media_y_list[i]:.4f}")
    print(f"  Desviacin Estndar eje X: {std_x_list[i]:.4f}")
    print(f"  Desviacin Estndar eje Y: {std_y_list[i]:.4f}")
    print(f"  Correlacin de ambos ejes: {correlation_list[i]:.4f}")
    print()
```
