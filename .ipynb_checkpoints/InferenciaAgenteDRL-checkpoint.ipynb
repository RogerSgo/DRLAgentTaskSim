```python
# Inferencia con ciclo anidado, for - while
# Inicializar librerias
import gymnasium   #
import EnvCGR_T   # Entorno personalizado
import stable_baselines3   # Biblioteca RL
from stable_baselines3 import A2C, PPO
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as np
# Inicializar el entorno
ENV_NAME = "EnvCGR_T/ConTask-v0"
env = gymnasium.make(ENV_NAME)
# Cargar modelo DRL
model = A2C.load("best_model", env=env)
val = []
X_ = []
Y_ = []
vx = []   # vector para posicion X
vy = []   # vector para posicion Y
vo = []   # vector para orientacion efector
vt = []   # vector orientacion detectada
vrdc = []
vrda = []
vrgc = []
vrat = []
vrft = []
ve = []

all_episode_rewards = []
num_episodes = 1
for i in range(num_episodes):
    episode_rewards = []
    done = False
    obs, info = env.reset()
    while not done:
        env.render()
        # states solo son tiles cuando se usan polticas LSTM
        action, states = model.predict(obs, deterministic=False)
        # Adqusicion de datos
        rt, rdc, rda, rgc, rat, rft, td = env.compute_reward(action)
        px = env.get_posicion(env.target)[0]
        py = env.get_posicion(env.target)[1]
        oe = env.get_orientacion(env.target)[2]
        
        vx.append(np.array(px))
        vy.append(np.array(py))
        vo.append(np.array(oe))
        vt.append(np.array(td))
        vrdc.append(np.array(rdc))
        vrda.append(np.array(rda))
        vrgc.append(np.array(rgc))
        vrat.append(np.array(rat))
        vrft.append(np.array(rft))
        ve.append(np.array(env.step_counter))
        
        # here, action, rewards and dones are arrays
        obs, reward, done, tru, info = env.step(action)
        episode_rewards.append(reward)
        val += [reward]
        X_.append(env.step_counter)
        Y_.append(np.array(val).mean())
        
    all_episode_rewards.append(sum(episode_rewards))
    
plt.figure(figsize=(4,2), dpi=500)
plt.plot(X_, Y_)
plt.grid()
plt.title('Acumulated Reward', fontsize=8)
plt.xlabel('Steps', fontsize=7)
plt.ylabel('Reward', fontsize=7)
plt.tick_params(axis='both', which='major', labelsize=7)
plt.show()

env.grafica(vx, vy)

mean_episode_reward = np.mean(all_episode_rewards)
print("Mean reward:", mean_episode_reward, "Num episodes:", num_episodes)

df = pd.DataFrame({'posX': vx, 'posY': vy, 'ori_efector': vo, 'ori_det': vt, 'rdc': vrdc, 
                   'rda': vrda, 'vrgc': vrgc, 'vrat': vrat, 'vrft': vrft, 'step': ve, 's': -0.025})
df.to_excel('Datos_DRLtt.xlsx', index=False, float_format='%.3f')

env.close()
```
